```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
area = np.array([[800],[1000],[1200],[1500],[1800]])
bedrooms = np.array([[1],[2],[2],[3],[3]])
house_price=[200000,300000,400000,500000,600000]
print("House data")
for i in range(len(area)):
    print(f"area: {area[i][0]} sqft, bedrooms: {bedrooms[i][0]}, price: ${house_price[i]}")
```

```python
x=np.column_stack((area,bedrooms))
y=np.array(house_price)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
print(model.coef_)
print(model.intercept_)
```
